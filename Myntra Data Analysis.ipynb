```python
import pandas as pd

df = pd.read_csv('products.csv')
df.head()
```

### 1. Checking Null Values

```python
df.isnull().sum()
```

### 2. Remove Columns that are not in use

```python
del df['product_link']
del df['img_link']
del df['meta_keyword']
del df['meta_description']

df.head()
```

### 3. Data Analysis - Numeric Column

###### Average rating of a product (non zero)

```python
rat = 0
c = 0

for i in df['rating']:
    
    if(i != 0.0):
        rat += i
        c += 1
        
print(round(int(rat) / c, 2))
```

##### 3.2) Most Expensive Product (Marked Price)

##### 3.3) Most Expensive Product (Discounted Price)

##### 3.5) Most Discounted Product (Disount Amount)

##### 3.6) Shortlisting Products based on specific filters

```python
df_pr = df[df['discounted_price'] > 2000]
df_pr = df_pr[df_pr['discounted_price'] < 3000]
df_pr = df_pr[df_pr['product_tag'] == 'handbags']


df_pr.sort_values(by = 'discount_percent', ascending = False).head(10)
```

#### 3.7) Finding all the brand names

```python
df['brand_name'].unique()
```

##### 3.8) Finding the top 10 Brands with most number of products

```python
data = []

for brand_name in df['brand_name'].unique():
    
    c = 0
    for i in df['brand_name']:
        if (i == brand_name):
            c += 1

    data.append([brand_name, c])
    
data = pd.DataFrame(data, columns = ['name','products'])

data.sort_values(by = 'products', ascending = False).head(10)
```

```python
import numpy as np

pincode = []
state   = []

for i in df['seller_address']:
    
    if (i != 'Not Mentioned'):
        
        pincode.append(i.split(' ')[-1])
        state.append(i.split(',')[-1].split('-')[0].strip())
        
    else:
        
        pincode.append(np.nan)
        state.append(np.nan)
        
df['pincode'] = pincode
df['state']   = state
```
